```python
import duckdb as ddb
import pandas as pd
import pathlib
```

```python
projdir = pathlib.Path().cwd()
pwdir = projdir / "passwords"
dbpath = projdir / "creads.db"
```

```python
conn = ddb.connect(dbpath)
conn.execute("CREATE OR REPLACE TABLE passwords (password VARCHAR UNIQUE)")
```

```python
q = """
INSERT INTO passwords (
  SELECT l.password 
  FROM _df l 
  ANTI JOIN passwords r 
    ON l.password = r.password
)
"""

for pwpath in pwdir.glob("*.txt"):
    _df = pd.read_fwf(pwpath, header=None, names=["password"]).drop_duplicates()    
    conn.sql(q)
```

```python
n_rows = conn.execute("SELECT count(*)FROM passwords;").fetchall()[0][0]
print(f"Anzahl Zeilen: {n_rows:_}")
# conn.close()
```
